```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
print(titanic.head())
```

countplot - means count the categorical or target class

```python

sns.countplot(x="survived", data=titanic)

plt.title("countplot of survived")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

Box Plot (Outlier Detection) - Shows median, quartiles, and outliers.

Used to detect abnormal values.

```python
data = titanic["age"].dropna()  

sns.boxplot(data=data)
plt.title("Box Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

Heatmap (Correlation Matrix) - it is mainly used to show the correlation between features.

Dark color → strong relationship
Light color → weak relationship

```python
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=False, cmap="coolwarm")
plt.show()
```

A violin plot shows: data distribution ,probability density,median and quartiles with class wise distribution

It is like a combination of box plot + distribution plot.

```python
sns.violinplot(x="survived", y="age", data=titanic)
plt.show()
```

Pair Plot(multivariate analysis) -  shows relationships between multiple variables at once.

Very useful during EDA.

```python
sns.pairplot(titanic, vars=["age","pclass","fare"], hue="survived")
plt.suptitle("Pair Plot")
plt.show()
```
