Line Plot - Used to show trend over time or sequence data.

Example: stock price, training loss.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,25,30]

plt.title("Line Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y)
plt.show()
```

Scatter Plot - Shows relationship between two variables.

Example: height vs weight.

Used for:

correlation
clustering visualization

```python
x = [1,2,3,4,5]
y = [5,7,8,9,10]

plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.scatter(x,y)
plt.show()
```

Bar Chart - Used to compare categories.

Example: sales of different products

```python
labels = ['A','B','C']
values = [10,20,15]

plt.title("Bar Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.bar(labels,values)
plt.show()
```

Histogram - Used to see data distribution.

Example: age distribution, marks distribution.

```python
data = [10,20,20,30,30,30,40]

plt.title("Histogram")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.hist(data,bins=5)
plt.show()
```

Pie Chart - Used to show percentage contribution.

```python
sizes = [40,30,20,100]
labels = ['A','B','C','D']

plt.title("Pie Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.pie(sizes, labels=labels)
plt.show()
```

legend() is used to show the label (name) of each plotted line or data series on the graph.When you plot multiple datasets, the legend helps you identify which line represents what.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
sales = [10,20,30,40]
profit = [5,10,15,20]

plt.plot(x, sales, label="Sales")
plt.plot(x, profit, label="Profit")

plt.legend()

plt.show()
```
