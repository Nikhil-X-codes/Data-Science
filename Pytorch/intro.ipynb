```python
import torch
import time

x = torch.randn(10000, 10000)

start = time.time()
y = x * x
print("CPU Time:", time.time() - start)

# GPU timing
device = "cuda" if torch.cuda.is_available() else "cpu"

x_gpu = x.to(device)

torch.cuda.synchronize()  

start = time.time()
y_gpu = x_gpu * x_gpu

torch.cuda.synchronize()   
print("GPU Time:", time.time() - start)
```
