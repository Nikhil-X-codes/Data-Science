Matrix multiplication of NUMPY

```python
W = [[1, 1, 1],
     [1.5, 1.5, 1.5],
     [2, 2, 2]]

x = [[6],
     [7],
     [8]]

b_np = np.ones((3, 1))

# y=Wx+b

output = np.matmul(W,x) + b_np

# Print out the result
print("Output:\n", output)
print("Output shape:\n", output.shape)
```

```python
W_torch = torch.FloatTensor(W)

x_torch = torch.FloatTensor(x)

b_torch = torch.ones(3, 1)

output = torch.matmul(W_torch, x_torch) + b_torch

print("Output:\n", output)
print("Output dtype:\n", output.dtype)
print("Output shape:\n", output.shape)
```

Default Datatype is Flot but we can convert into int

```python
W_torch = torch.FloatTensor(W).to(torch.int64)

x_torch = torch.FloatTensor(x).to(torch.int64)

b_torch = torch.ones(3, 1).to(torch.int64)

output = torch.matmul(W_torch, x_torch) + b_torch

print("Output:\n", output)
print("Output dtype:\n", output.dtype)
print("Output shape:\n", output.shape)
```

torch.matmul() - General-purpose matrix multiplication (1-D,2-D)

torch.bmm() - batch matrix multiplication (3-D)

```python
tensor_1 = torch.rand(3,3,3)
tensor_2 = torch.rand(3,3,3)

print("Tensor 1:\n",tensor_1)
print("Tensor 2:\n",tensor_2)

print("Addition:\n",tensor_1 + tensor_2)

print("Batch Multiplication:\n", torch.bmm(tensor_1, tensor_2))
```

```python
print("The Tensor's shape is:", tensor_1.shape)

print("The Tensor's shape using size() is:", tensor_1.size())

print("There are %d elements in total:" % tensor_1.numel())

print("There are %d Dimensions" %(tensor_1.ndim))

print("Flattened Tensor:\n", tensor_1.flatten())
```

Autograd (Automatic Differentiation) 

Why Gradients Matter

Neural networks learn by updating weights.

To update weights, we need:
gradient of loss with respect to weights and Mathematically:

∂W
∂Loss
	​

Doing this manually for large networks is extremely difficult.

```python
x = torch.FloatTensor([4])
x.requires_grad = True
w = torch.FloatTensor([2])
w.requires_grad = True
b = torch.FloatTensor([3])
b.requires_grad = True

# By performing a simple computation Pytorch will build a computational graph.
y = w * x + b   


y.backward()


print("Calculated Gradients") 
print("dy/dx", x.grad.item())    # x.grad = dy/dx = 2 
print("dy/dw", w.grad.item())    # w.grad = dy/dw = 4
print("dy/db", b.grad.item())   # b.grad = dy/db = 1  
```
